```python
import time
import sys
import os

# Fungsi untuk buat teks keluar satu-satunya (Typing effect)
def taip(teks, kelajuan=0.04):
    for huruf in teks:
        sys.stdout.write(huruf)
        sys.stdout.flush()
        time.sleep(kelajuan)
    print()

# Fungsi animasi loading sempoi
def animasi_loading():
    os.system('cls' if os.name == 'nt' else 'clear')
    print("\n\n")
    for i in range(3):
        for lambang in ["  ( •_•)", "  ( •_•)>⌐■-■", "  (⌐■_■)"]:
            sys.stdout.write(f"\r  {lambang}  Memuatkan Game Teka-Teki... ")
            sys.stdout.flush()
            time.sleep(0.3)
    os.system('cls' if os.name == 'nt' else 'clear')

# Sound effect guna system beep (Windows sahaja, kalau Mac/Linux dia senyap)
def bunyi_betul():
    if os.name == 'nt':
        import winsound
        winsound.Beep(1000, 200)
        winsound.Beep(1500, 300)

def bunyi_salah():
    if os.name == 'nt':
        import winsound
        winsound.Beep(400, 500)

# Animasi bila MENANG gempak
def animasi_menang():
    os.system('cls' if os.name == 'nt' else 'clear')
    bunyi_betul()
    piala = """
     _______________________
    |  ___________________  |
    | |                   | |
    | |   AMAZING!!!      | |
    | |   HAMPA MENANG!   | |
    | |___________________| |
    |_______________________|
           \\       /
            \\     /
             |   |
            /     \\
           /_______\_
    """
    print(piala)
    taip("  🎉 TAHNIAH! KELAS HAMPA MEMANG MEMANG SEMPOI DAN BIJAK! 🎉\n", 0.05)

# Animasi bila KALAH
def animasi_kalah():
    os.system('cls' if os.name == 'nt' else 'clear')
    bunyi_salah()
    tengkorak = """
         ______
      .-"      "-.
     /            \\
    |              |
    |,  .-.  .-.  ,|
    | )(__/  \\__)( |
    |/     /\\     \\|
    (_     ^^     _)
     \\__|IIIIII|__/
      | \\IIIIII/ |
      \\          /
       `--------`
    """
    print(tengkorak)
    taip("  💥 GAME OVER... Sayang sekali, cuba lagi nanti! 💥\n", 0.05)

# --- MAIN GAME ---
def main_game():
    animasi_loading()

    print("="*60)
    taip("⚡ SELAMAT DATANG KE GAME TEKA-TEKI ULTRA-FUN ⚡", 0.03)
    print("="*60)
    print("\n")

    # Senarai soalan, pilihan jawapan (A/B/C), dan jawapan yang betul
    soalan_list = [
        {
            "soalan": "1. Banyak-banyak mi, mi apa yang kalau kita picit, dia keluar bunyi?",
            "pilihan": "A. Mi Goreng\nB. Mi Sup\nC. Miscall",
            "jawapan": "C"
        },
        {
            "soalan": "2. Ada sebuah bot lumba, dalam bot tu ada seekor monyet, seekor burung, dan seekor kucing. Bot tu karam, siapa yang selamat?",
            "pilihan": "A. Monyet sebab pandai berenang\nB. Burung sebab boleh terbang\nC. Kucing sebab ada 9 nyawa\nD. Semua mati sebab bot tu bot lumba, bukan bot penyelamat",
            "jawapan": "B"
        },
        {
            "soalan": "3. Dia ada 4 kaki pada waktu pagi, 2 kaki pada waktu tengahari, dan 3 kaki pada waktu malam. Siapakah dia?",
            "pilihan": "A. Raksasa\nB. Manusia (Bayi -> Dewasa -> Tua pakai tongkat)\nC. Meja lipat",
            "jawapan": "B"
        }
    ]

    skor = 0
    nyawa = 3

    for i, q in enumerate(soalan_list):
        if nyawa <= 0:
            break

        print("-" * 50)
        taip(q["soalan"], 0.04)
        print(q["pilihan"])
        print("-" * 50)

        # Minta input dari student
        jawapan_user = input("Jawapan hampa (Tulis huruf SAHAJA): ").strip().upper()

        if jawapan_user == q["jawapan"]:
            print("\n")
            taip("✨ BETULLL! Mantap la hangpa! ✨", 0.02)
            if os.name == 'nt':
                import winsound
                winsound.Beep(1200, 150)
            skor += 1
            time.sleep(1)
        else:
            print("\n")
            taip("❌ SALAHHH! Aduhhh... ❌", 0.02)
            if os.name == 'nt':
                import winsound
                winsound.Beep(300, 400)
            nyawa -= 1
            print(f"Nyawa tinggal: {nyawa} ❤️")
            time.sleep(1)

        os.system('cls' if os.name == 'nt' else 'clear')

    # Keputusan Akhir
    if skor == len(soalan_list):
        animasi_menang()
        print(f"Skor Penuh: {skor}/{len(soalan_list)}")
    else:
        animasi_kalah()
        print(f"Skor Akhir: {skor}/{len(soalan_list)}")

# Jalankan game
if __name__ == "__main__":
    main_game()
```
